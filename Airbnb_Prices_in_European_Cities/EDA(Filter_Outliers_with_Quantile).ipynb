```python
def filter_outliers(mask_column):
  global dataframe
  low, high = dataframe[mask_column].quantile([0.0, 0.9])
  mask_column = dataframe[mask_column].between(low, high)
  dataframe = dataframe[mask_column]
```

```python
mask_Preço = "Preço_EURO"
filter_outliers(mask_Preço)
```
