```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
dataframe = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/Datasets/Airbnb_Prices_in_European_Cities/cities_weekdays.csv", index_col=[0])
```

# **Verificando Preço_EURO**

```python
plt.hist(dataframe["Preço_EURO"]);
```

```python
dataframe["Preço_EURO"].describe()
```

```python
stats.shapiro(dataframe["Preço_EURO"])
```

```python
sm.qqplot(dataframe["Preço_EURO"], line='s')
plt.show()
```

```python
def filter_outliers(mask_column):
  global dataframe
  low, high = dataframe[mask_column].quantile([0.0, 0.9])
  mask_column = dataframe[mask_column].between(low, high)
  dataframe = dataframe[mask_column]
```

```python
mask_Preço = "Preço_EURO"
filter_outliers(mask_Preço)
```

```python
plt.hist(dataframe["Preço_EURO"]);
```

```python
dataframe["Preço_EURO"].describe()
```

```python
stats.shapiro(dataframe["Preço_EURO"])
```

```python
sm.qqplot(dataframe["Preço_EURO"], line='s')
plt.show()
```
