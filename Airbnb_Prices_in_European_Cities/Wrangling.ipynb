```python
import pandas as pd
from glob import glob
```

```python
def wrangle(city, file):
  df = pd.read_csv(file)
  df.drop(columns=[
      "Unnamed: 0", "room_private", "room_shared", "biz", "bedrooms", "multi", "attr_index", "rest_index",
    ], 
      inplace= True)
  df.rename(columns={
      "realSum": "Preço_EURO", # o preço total para 2 pessoas e 2 noites (em EURO)
      "room_type": "Tipo_Quarto", 
      "person_capacity": "Máximo_Pessoas",
      "host_is_superhost": "SuperHost?",
      "cleanliness_rating": "Avaliação_Limpeza",
      "guest_satisfaction_overall": "Avaliação_Acomodação",
      "dist": "Dist.Centro", # distância do Centro da Cidade em KM
      "metro_dist": "Dist.Metrô", # distância da Estação de Metrô (mais próxima) em KM
      "attr_index_norm": "Índice_Atração", # índice de atrações próximos
      "rest_index_norm": "Índice_Restaurantes", # índice de restaurantes próximos
      "lng": "Lon",
      "lat": "Lat"
    }, 
      inplace=True)
  df["Preço_EURO"] = df["Preço_EURO"].round(2)
  df["Avaliação_Acomodação"] = df["Avaliação_Acomodação"] / 10
  df["Máximo_Pessoas"] = df["Máximo_Pessoas"].astype(int)
  df["Cidade"] = city

  return df
```

```python
files = glob("/content/drive/MyDrive/Colab_Notebooks/Datasets/Airbnb_Prices_in_European_Cities/*_weekdays.csv")
```

```python
dataframes = [] 
for file in files:
  city = file.split("/")[7].split("_")[0]
  file = wrangle(city, file)
  dataframes.append(file)
```

```python
df = pd.concat(dataframes, ignore_index=True)
```
