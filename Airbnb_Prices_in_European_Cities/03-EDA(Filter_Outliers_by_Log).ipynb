```python
import pandas as pd
import matplotlib.pyplot as plt
from math import log
from scipy import stats
import statsmodels.api as sm
```

```python
dataframe = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/Datasets/Airbnb_Prices_in_European_Cities/cities_weekdays.csv", index_col=[0])
```

```python
dataframe.head()
```

```python
dataframe.shape
```

# **Verificando Preço_EURO**

```python
plt.hist(dataframe["Preço_EURO"]);
```

```python
dataframe["Preço_EURO"].describe()
```

```python
stats.shapiro(dataframe["Preço_EURO"])
```

```python
sm.qqplot(dataframe["Preço_EURO"], line='s')
plt.show()
```

```python
def cal_log(x):
  y = log(x, 10)
  y = round(y, 2)
  return y
```

```python
log_Preço = []
for i in dataframe["Preço_EURO"]:
  results = cal_log(i)
  log_Preço.append(results)
```

```python
dataframe["log_Preço_EURO"] = log_Preço
```

```python
plt.hist(dataframe["log_Preço_EURO"]);
```

```python
dataframe["log_Preço_EURO"].describe()
```

```python
stats.shapiro(dataframe["log_Preço_EURO"])
```

```python
sm.qqplot(dataframe["log_Preço_EURO"], line='s')
plt.show()
```
